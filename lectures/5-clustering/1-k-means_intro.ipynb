# K-Means Clustering

Based on [this](https://blog.floydhub.com/introduction-to-k-means-clustering-in-python-with-scikit-learn/).

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Small 2D set

```python
# generate a small 2D dataset
X, y = make_blobs(n_samples=10, centers=3, cluster_std=0.60, random_state=0)
```

```python
# plot the data set
plt.scatter(X[:,0], X[:,1])
```

```python
# perform k-means clustering    
# specify the number of clusters (3) and fit the data X
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
# get the cluster centroids
print(kmeans.cluster_centers_)
```

```python
# get the cluster labels
print(kmeans.labels_)
```

```python
# plot the cluster centers and the data points
plt.scatter(X[:, 0], X[:, -1])
    
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x')
    
plt.title('Data points and cluster centroids')
plt.show()
```

## Bigger 2D set

```python
# generate a bigger 2D data set
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
# plot it
plt.scatter(X[:,0], X[:,1])
```

```python
# specify the number of clusters (4) and fit the data X
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
```

```python
# plot the cluster centers and the data points on a 2D plane
plt.scatter(X[:, 0], X[:, -1])
    
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x')
    
plt.title('Data points and cluster centroids')
plt.show()
```
